# Create a Rasterized Grid of California Coordinates

```python
import pandas as pd
import numpy as np
import pickle
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
from geocube.api.core import make_geocube
pd.options.mode.chained_assignment = None  # default='warn' (hush dumb pandas warnings)
```

```python
usa = gpd.read_file('../us_wildfire_dataset/states.shp')
CA = usa[usa.STATE_ABBR == 'CA']
CA.plot()
```

```python
CA["exists"] = 1  # binary in-california
raster = make_geocube(
    vector_data=CA,
    measurements=["exists"],
    resolution=(0.05, -0.05),
    fill=np.nan,
).fillna(0)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots()
raster.where(raster!=0).exists.plot(ax=axes)
```

```python
with open("../us_wildfire_dataset/ca_raster_grid.pkl", "wb") as f:
    pickle.dump(raster, f, protocol=pickle.HIGHEST_PROTOCOL)
```
